```python
import pandas as pd
```

```python
ne = pd.read_csv('/content/gdp_nebraska_data_final.csv')
nh = pd.read_csv('/content/gdp_new_hampshire_data_final.csv')
wv = pd.read_csv('/content/gdp_west_virginia_data_final.csv')
```

```python
ne
```

```python
# melt each dataframe to long format and add a 'State' column
ne_melted = ne.melt(id_vars=['GeoName'], var_name='Year', value_name='GDP (USD)')
ne_melted['State'] = 'Nebraska'

nh_melted = nh.melt(id_vars=['GeoName'], var_name='Year', value_name='GDP (USD)')
nh_melted['State'] = 'New Hampshire'

wv_melted = wv.melt(id_vars=['GeoName'], var_name='Year', value_name='GDP (USD)')
wv_melted['State'] = 'West Virginia'

# concatenate the melted dataframes
combined_data = pd.concat([ne_melted, nh_melted, wv_melted], ignore_index=True)

combined_data = combined_data.rename(columns={'GeoName': 'County'})

state_order = ['Nebraska', 'New Hampshire', 'West Virginia']
combined_data['State'] = pd.Categorical(combined_data['State'], categories=state_order, ordered=True)
combined_data = combined_data.sort_values(by=['State', 'Year', 'County']).reset_index(drop=True)

# reorder columns
combined_data = combined_data[['Year', 'State', 'County', 'GDP (USD)']]

combined_data
```

```python
combined_data.to_csv('gdp_data.csv', index=False)
```
