```python
import pandas as pd
```

```python
drug_overdose_death_df = pd.read_csv("/content/VSRR_Provisional_County-Level_Drug_Overdose_Death_Counts.csv")
drug_overdose_death_df
```

```python
drug_overdose_death_df.fillna(0, inplace=True)
drug_overdose_death_df = drug_overdose_death_df[drug_overdose_death_df['STATE_NAME'].isin(['West Virginia', 'New Hampshire', 'Nebraska'])]
drug_overdose_death_df
```

```python
drug_overdose_death_df['Provisional Drug Overdose Deaths'] = pd.to_numeric(drug_overdose_death_df['Provisional Drug Overdose Deaths'], errors='coerce')
drug_overdose_death_df = drug_overdose_death_df.groupby(['Year', 'STATE_NAME', 'COUNTYNAME'])['Provisional Drug Overdose Deaths'].sum().reset_index()
drug_overdose_death_df
```

```python
drug_overdose_death_df = drug_overdose_death_df.rename(columns={
    'Provisional Drug Overdose Deaths': 'Drug Overdose Deaths',
    'STATE_NAME': 'State',
    'COUNTYNAME': 'County'
    })
drug_overdose_death_df = drug_overdose_death_df.sort_values(by=['Year', 'State']).reset_index(drop=True)
drug_overdose_death_df
```

```python
import pandas as pd
state_order = ['Nebraska', 'New Hampshire', 'West Virginia']
drug_overdose_death_df['State'] = pd.Categorical(drug_overdose_death_df['State'], categories=state_order, ordered=True)
drug_overdose_death_df = drug_overdose_death_df.sort_values(by=['State', 'Year', 'County']).reset_index(drop=True)
drug_overdose_death_df = drug_overdose_death_df[drug_overdose_death_df['Year'] != 2024]
drug_overdose_death_df
```

```python
drug_overdose_death_df.to_csv('drug_overdose_death_data.csv', index=False)
```
