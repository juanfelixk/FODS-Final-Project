```python
import pandas as pd
```

```python
data = pd.read_csv('/content/County Opioid Dispensing Rates.csv')
data
```

```python
data = data[['YEAR', 'STATE_NAME', 'COUNTY_NAME', 'opioid_dispensing_rate']]
data = data[data['STATE_NAME'].isin(['West Virginia', 'New Hampshire', 'Nebraska'])].reset_index(drop=True)
data = data[data['YEAR'] != 2019]
data
```

```python
data = data.rename(columns={
    'YEAR': 'Year',
    'STATE_NAME': 'State',
    'COUNTY_NAME': 'County',
    'opioid_dispensing_rate': 'Opioid Dispensing Rate (per 100 persons)'
})

state_order = ['Nebraska', 'New Hampshire', 'West Virginia']
data['State'] = pd.Categorical(data['State'], categories=state_order, ordered=True)
data = data.sort_values(by=['State', 'Year', 'County']).reset_index(drop=True)
data
```

```python
import numpy as np

data['Opioid Dispensing Rate (per 100 persons)'] = data['Opioid Dispensing Rate (per 100 persons)'].replace('Data unavailable', np.nan)
data
```

```python
data['Opioid Dispensing Rate (per 100 persons)'] = pd.to_numeric(data['Opioid Dispensing Rate (per 100 persons)'], errors='coerce')
data
```

```python
data.to_csv('opioid_dispensing_rate_data.csv', index=False)
```
