```python
import pandas as pd
```

```python
data_2020 = pd.read_csv('/content/poverty 2020.csv', header=3)
data_2021 = pd.read_csv('/content/poverty 2021.csv', header=3)
data_2022 = pd.read_csv('/content/poverty 2022.csv', header=3)
data_2023 = pd.read_csv('/content/poverty 2023.csv', header=3)
```

```python
data_2020
```

```python
data_2020 = data_2020[data_2020['Postal Code'].isin(['NE', 'NH', 'WV'])]
data_2020 = data_2020[data_2020['County FIPS Code'] != 0]
data_2020 = data_2020[['Postal Code', 'Name', 'Poverty Estimate, All Ages']].reset_index(drop=True)
data_2020.rename(
    columns = {
        'Postal Code': 'State',
        'Name': 'County',
        'Poverty Estimate, All Ages': 'Poverty Estimate'
    }, inplace=True
)
data_2020['State'] = data_2020['State'].replace(
    {
      'NE': 'Nebraska',
      'NH': 'New Hampshire',
      'WV': 'West Virginia'
    }
)
data_2020['Year'] = 2020
data_2020 = data_2020[['Year', 'State', 'County', 'Poverty Estimate']]
data_2020
```

```python
data_2021 = data_2021[data_2021['Postal Code'].isin(['NE', 'NH', 'WV'])]
data_2021 = data_2021[data_2021['County FIPS Code'] != 0]
data_2021 = data_2021[['Postal Code', 'Name', 'Poverty Estimate, All Ages']].reset_index(drop=True)
data_2021.rename(
    columns = {
        'Postal Code': 'State',
        'Name': 'County',
        'Poverty Estimate, All Ages': 'Poverty Estimate'
    }, inplace=True
)
data_2021['State'] = data_2021['State'].replace(
    {
      'NE': 'Nebraska',
      'NH': 'New Hampshire',
      'WV': 'West Virginia'
    }
)
data_2021['Year'] = 2021
data_2021 = data_2021[['Year', 'State', 'County', 'Poverty Estimate']]
data_2021
```

```python
data_2022 = data_2022[data_2022['Postal Code'].isin(['NE', 'NH', 'WV'])]
data_2022 = data_2022[data_2022['County FIPS Code'] != 0]
data_2022 = data_2022[['Postal Code', 'Name', 'Poverty Estimate, All Ages']].reset_index(drop=True)
data_2022.rename(
    columns = {
        'Postal Code': 'State',
        'Name': 'County',
        'Poverty Estimate, All Ages': 'Poverty Estimate'
    }, inplace=True
)
data_2022['State'] = data_2022['State'].replace(
    {
      'NE': 'Nebraska',
      'NH': 'New Hampshire',
      'WV': 'West Virginia'
    }
)
data_2022['Year'] = 2022
data_2022 = data_2022[['Year', 'State', 'County', 'Poverty Estimate']]
data_2022
```

```python
data_2023 = data_2023[data_2023['Postal Code'].isin(['NE', 'NH', 'WV'])]
data_2023 = data_2023[data_2023['County FIPS Code'] != 0]
data_2023 = data_2023[['Postal Code', 'Name', 'Poverty Estimate, All Ages']].reset_index(drop=True)
data_2023.rename(
    columns = {
        'Postal Code': 'State',
        'Name': 'County',
        'Poverty Estimate, All Ages': 'Poverty Estimate'
    }, inplace=True
)
data_2023['State'] = data_2023['State'].replace(
    {
      'NE': 'Nebraska',
      'NH': 'New Hampshire',
      'WV': 'West Virginia'
    }
)
data_2023['Year'] = 2023
data_2023 = data_2023[['Year', 'State', 'County', 'Poverty Estimate']]
data_2023
```

```python
merged_data = pd.concat([data_2020, data_2021, data_2022, data_2023], ignore_index=True)
merged_data
```

```python
merged_data = merged_data.sort_values(by='Year').reset_index(drop=True)
merged_data.head()
```

```python
merged_data.to_csv('poverty_data.csv', index=False)
```
