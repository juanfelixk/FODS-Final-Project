```python
import pandas as pd
```

```python
data_2023 = pd.read_csv('sahie_2023.csv', skiprows=84)
data_2023 = data_2023[(data_2023['countyfips'] != 0) & (data_2023['geocat'] == 50) & (data_2023['agecat'] == 0) & (data_2023['sexcat'] == 0) & (data_2023['iprcat'] == 0)]
data_2023['state_name'] = data_2023['state_name'].str.strip()
data_2023['county_name'] = data_2023['county_name'].str.strip().str.replace(' County', '')
data_2023 = data_2023[data_2023['state_name'].isin(['West Virginia','Nebraska','New Hampshire'])]
data_2023 = data_2023[['year', 'PCTELIG', 'state_name', 'county_name']].reset_index(drop=True)
data_2023 = data_2023.rename(columns={
    'year': 'Year',
    'PCTELIG': 'Uninsured (%)',
    'state_name': 'State',
    'county_name': 'County'
})
data_2023
```

```python
data_2022 = pd.read_csv('sahie_2022.csv', skiprows=84)
data_2022 = data_2022[(data_2022['countyfips'] != 0) & (data_2022['geocat'] == 50) & (data_2022['agecat'] == 0) & (data_2022['sexcat'] == 0) & (data_2022['iprcat'] == 0)]
data_2022['state_name'] = data_2022['state_name'].str.strip()
data_2022['county_name'] = data_2022['county_name'].str.strip().str.replace(' County', '')
data_2022 = data_2022[data_2022['state_name'].isin(['West Virginia','Nebraska','New Hampshire'])]
data_2022 = data_2022[['year', 'PCTELIG', 'state_name', 'county_name']].reset_index(drop=True)
data_2022 = data_2022.rename(columns={
    'year': 'Year',
    'PCTELIG': 'Uninsured (%)',
    'state_name': 'State',
    'county_name': 'County'
})
data_2022
```

```python
data_2021 = pd.read_csv('sahie_2021.csv', skiprows=83)
data_2021 = data_2021[(data_2021['countyfips'] != 0) & (data_2021['geocat'] == 50) & (data_2021['agecat'] == 0) & (data_2021['sexcat'] == 0) & (data_2021['iprcat'] == 0)].copy()
data_2021['state_name'] = data_2021['state_name'].str.strip()
data_2021['county_name'] = data_2021['county_name'].str.strip().str.replace(' County', '')
data_2021 = data_2021[data_2021['state_name'].isin(['West Virginia','Nebraska','New Hampshire'])]
data_2021 = data_2021[['year', 'PCTELIG', 'state_name', 'county_name']].reset_index(drop=True)
data_2021 = data_2021.rename(columns={
    'year': 'Year',
    'PCTELIG': 'Uninsured (%)',
    'state_name': 'State',
    'county_name': 'County'
})
data_2021
```

```python
data_2020 = pd.read_csv('sahie_2020.csv', skiprows=79)
data_2020 = data_2020[(data_2020['countyfips'] != 0) & (data_2020['geocat'] == 50) & (data_2020['agecat'] == 0) & (data_2020['sexcat'] == 0) & (data_2020['iprcat'] == 0)].copy()
data_2020['state_name'] = data_2020['state_name'].str.strip()
data_2020['county_name'] = data_2020['county_name'].str.strip().str.replace(' County', '')
data_2020 = data_2020[data_2020['state_name'].isin(['West Virginia','Nebraska','New Hampshire'])]
data_2020 = data_2020[['year', 'PCTELIG', 'state_name', 'county_name']].reset_index(drop=True)
data_2020 = data_2020.rename(columns={
    'year': 'Year',
    'PCTELIG': 'Uninsured (%)',
    'state_name': 'State',
    'county_name': 'County'
})
data_2020
```

```python
combined_data = pd.concat([data_2020, data_2021, data_2022, data_2023], ignore_index=True)

# reorder columns to move Uninsured (%) to the right end
cols = list(combined_data.columns)
cols.remove('Uninsured (%)')
combined_data = combined_data[cols + ['Uninsured (%)']]
combined_data
```

```python
state_order = ['Nebraska', 'New Hampshire', 'West Virginia']
combined_data['State'] = pd.Categorical(combined_data['State'], categories=state_order, ordered=True)
combined_data = combined_data.sort_values(by=['State', 'Year', 'County']).reset_index(drop=True)
combined_data
```

```python
combined_data.to_csv('health_insurance_data.csv', index=False)
```
