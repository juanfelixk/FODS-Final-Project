```python
import pandas as pd
```

```python
df = pd.read_csv('/content/unemployment (1).csv')
```

```python
state_order = ['Nebraska', 'New Hampshire', 'West Virginia']
df['State'] = pd.Categorical(df['State'], categories=state_order, ordered=True)
df = df.sort_values(by=['State', 'Year', 'County']).reset_index(drop=True)
df
```

```python
# Count unique counties per state
df.groupby("State")["County"].nunique()

# Check for duplicate county-year pairs
duplicates = df[df.duplicated(subset=["State", "County", "Year"], keep=False)]

# Find extra county names (discrepancies) not in the official list
official_counts = {"Nebraska": 93, "West Virginia": 55, "New Hampshire": 10}
df.groupby("State")["County"].nunique() - pd.Series(official_counts)
```

```python
reference = pd.read_csv('/content/gdp_data.csv')

unique_counties_df = set(df['County'].unique())
unique_counties_reference = set(reference['County'].unique())

counties_not_in_reference = unique_counties_df - unique_counties_reference

print(f"Number of counties in unemployment data not found in reference data: {len(counties_not_in_reference)}")
print("Counties not in reference data:")
for county in sorted(list(counties_not_in_reference)):
    print(county)
```

```python
counties_to_remove = [
    'Bellevue',
    'Charleston',
    'Elkins',
    'Fairmont',
    'Grand Island',
    'Hastings',
    'Huntington',
    'Lexington',
    'Norfolk',
    'Parkerson',
    'Wheeling'
]

print(f"Shape of DataFrame before removing the 11 counties: {df.shape}")
df = df[~df['County'].isin(counties_to_remove)]
print(f"Shape of DataFrame after removing the 11 counties: {df.shape}")

display(df.head())
```
